```python
# Import pandas
import pandas as pd
```

```python
tetons_weather_url = (
    'https://www.ncei.noaa.gov/access/services/data/v1?'
    'dataset=daily-summaries'
    '&dataTypes=TOBS,PRCP'
    '&stations=USC00486428'
    '&startDate=1970-10-01&endDate=2024-05-03'
    '&includeStationName=true'
    '&includeStationLocation=1'
    '&units=standard')
tetons_weather_url
```

```python
tetons_df = pd.read_csv(
    tetons_weather_url, index_col="DATE", parse_dates=True, na_values=["NaN"])
tetons_df
```

```python
# Check that the data was imported into a pandas DataFrame
type(tetons_df)
```

```python
tetons_df = tetons_df[['PRCP', 'TOBS']]
tetons_df
```

Deleted excess info from table; table now shows date, observed temperature (TOBS) and precipitation (PRCP)

```python
tetons_df.plot()
```

Generated plot from data with no labels or parameters.

```python
# Plot the data using .plot
tetons_df.plot(
    y='TOBS',
    title='Moose, WY Daily Temperature 1970-2024',
    xlabel='Year',
    ylabel='Temperature (F)',
    legend=False,
    color='orange',
    figsize=(10,5))
tetons_df 
```
